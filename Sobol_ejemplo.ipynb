```python
# Importamos las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.qmc import Sobol

# Definimos el rango de búsqueda para los hiperparámetros
learning_rate_bounds = [0.01, 0.1]  # Rango para learning rate
batch_size_bounds = [16, 128]       # Rango para batch size

# Número de muestras a generar
n_samples = 20

# Creamos un generador Sobol para un espacio de 2 dimensiones (learning rate y batch size)
sobol_sampler = Sobol(d=2, scramble=False)

# Generamos las muestras en un espacio normalizado [0, 1]
samples = sobol_sampler.random(n=n_samples)

# Reescalamos las muestras al rango de los hiperparámetros
samples[:, 0] = samples[:, 0] * (learning_rate_bounds[1] - learning_rate_bounds[0]) + learning_rate_bounds[0]
samples[:, 1] = samples[:, 1] * (batch_size_bounds[1] - batch_size_bounds[0]) + batch_size_bounds[0]

# Graficamos los puntos generados
plt.figure(figsize=(8, 6))
plt.scatter(samples[:, 0], samples[:, 1], color='blue', label="Sobol Sampling Points")
plt.title("Sobol Sampling for Hyperparameter Optimization", fontsize=14)
plt.xlabel("Learning Rate", fontsize=12)
plt.ylabel("Batch Size", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
